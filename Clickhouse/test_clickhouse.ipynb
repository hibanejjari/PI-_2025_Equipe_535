```python
import clickhouse_connect
```

```python
# Create ClickHouse client
client = clickhouse_connect.get_client(
    host='localhost',      # or ClickHouse server IP / hostname
    port=8123,             # default HTTP port 9000 TCP port par défaut <- plus simple pour l'ingestion de données
    username='default',
    password='',
    database='default'
)
```

```python
# Run a query
query = """
SELECT
    user_id,
    event_time,
    event_type
FROM events
LIMIT 100
"""
```

```python
# Fetch results
result = client.query(query)

# Data as Python objects
rows = result.result_rows
columns = result.column_names

print(columns)
for row in rows[:5]:
    print(row)
```
